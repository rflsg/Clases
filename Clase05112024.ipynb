```python
import matplotlib.pyplot as plt
import numpy as np

font1={'family':'serif','color':'blue','size':20}

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)
plt.title('Primera Gráfica', fontdict=font1)
plt.xlabel('x')
plt.ylabel('y')

plt.grid(axis='y')


#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y, linestyle = 'dotted')
plt.title('Segunda Gráfica', fontdict=font1)
plt.xlabel('x')
plt.ylabel('y')

plt.grid(axis='x')

plt.show()
```
