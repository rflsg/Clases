```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t2 = np.arange(0.0, 5.0, 0.1)

plt.subplot(1,2,1)
plt.plot(t2, np.cos(2*np.pi*t2))


#plot 2:
def g(t):
    return np.sin(t) * 2

t1= np.arange(0.0, 50.0, 0.2)

plt.subplot(1, 2, 2)
plt.plot(g(t1))
plt.title('Grafica 2')

plt.show()
```
